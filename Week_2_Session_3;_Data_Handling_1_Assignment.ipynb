**Part 2: Column Selection & Basic Filtering**


A. Selecting Columns

```python
##Selecting any two numerical columns and display them
data_sales.loc[:, ["MONTH",	"YEAR"]]
```

```python
#Select any two categorical columns and display them.
data_sales.loc[:, ["SUPPLIER",	"ITEM TYPE"]]
```

**B. Filtering Rows**

```python
#filtering rows using one condition
data_sales[data_sales['WAREHOUSE SALES']> 25]
```

```python
#Filter rows using two conditions (using & or |)
data_sales.loc[(data_sales['WAREHOUSE SALES'] > 25) & (data_sales['RETAIL SALES'] >= 30)]
```

```python
#Display only 23 selected columns for those rows.
data_sales.loc[(data_sales['WAREHOUSE SALES'] > 25) & (data_sales['RETAIL SALES'] >= 30), data_sales.columns[:3]]
```

```python
#The first 5 rows and first 3 columns#
data_sales.iloc[:5, :3]
```

```python
#Rows 1020 and columns 24#
data_sales.iloc[10:21,1:5]
```

```python
data_sales.groupby("ITEM DESCRIPTION").size()
```

```python
#Calculate the total of a numerical column for each category.#
data_sales.groupby("ITEM DESCRIPTION")["RETAIL SALES"].sum()
```

```python
data_sales.groupby("ITEM DESCRIPTION").agg({
    "RETAIL SALES":['sum','mean','max','count'],
    "MONTH":['sum','mean','max','count'],
    "ITEM CODE":['count'] # 'mean', 'sum', and 'max' are not meaningful for item codes#
    })
```
