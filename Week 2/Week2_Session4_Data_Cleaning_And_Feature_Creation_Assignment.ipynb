Cleaning of the Data

```python
# Decision: Drop this column, IT IS NOT Needed for analysis
sales_data_clean = sales_data.drop(columns=['ITEM DESCRIPTION'])
```

For the na in the column of item type, i would add it to the  most occuring count of the unique.


```python
unique_count = sales_data_clean["ITEM TYPE"].value_counts()
print("Number of unique values:", unique_count)
```

```python
# FILLING THE MISSING ROW UNDER ITEM TYPE WITH WINE
sales_data_clean['ITEM TYPE'] = sales_data_clean['ITEM TYPE'].fillna('WINE')
```

```python
sales_data_clean.head()
```

```python
# FILL MISSING VALUES UNDER THE COLUMN 'SUPPLIER'
sales_data_clean['SUPPLIER'] = sales_data_clean['SUPPLIER'].fillna('UNKNOWN')
```

```python
# FILL THE MISSING VALUES UNDER RETAIL SALES WITH THE MODE
sales_data_clean['RETAIL SALES'] = sales_data_clean['RETAIL SALES'].fillna(sales_data_clean['RETAIL SALES'].mode()[0])
```

CHECK IF THERE ARE STILL MISSING VALUES

```python
sales_data_clean.isna().sum()
```

CHECK FOR DUPLICATES

```python
sales_data_clean.duplicated().sum()
```

FEATURE ENGINEERING

```python
# CREATE TOTAL SALES COLUMN TO SEE TOTAL SALES FROM EACH TRANSACTIONS
sales_data_clean['TOTAL SALES'] = sales_data_clean['RETAIL SALES'] + sales_data_clean['WAREHOUSE SALES']
sales_data_clean.head()
```

```python
# CREATE A NEW COLUMN 'DATE' AND MAKE IT A DATETIME DATATYPE
sales_data_clean['DATE'] = sales_data_clean['YEAR'].astype(str) + '-' + sales_data_clean['MONTH'].astype(str)
sales_data_clean['DATE'] = pd.to_datetime(sales_data_clean['DATE'])
sales_data_clean.info()
```
